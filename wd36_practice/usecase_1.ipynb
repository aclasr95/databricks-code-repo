**Read the file and create a dataframe**

```python
usage_df = spark.read.option("header", "true").format("csv").load("dbfs:/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv")
usage_df.display()
```

_Wrting the created DF into a table **default.mobile_os_usage'**_

```python
#usage_df.write.mode("overwrite").saveAsTable("default.mobile_os_usage")
usage_df = usage_df.withColumnRenamed(
    "Mobile Operating System", "Mobile_Operating_System"
).withColumnRenamed(
    "Percent of Usage", "Percent_of_Usage"
)
usage_df.write.mode("overwrite").saveAsTable("default.mobile_os_usage")

print("completed")
```

**Sql query to fetch the data from the above table created**

```python
%sql
select * from default.mobile_os_usage
```
